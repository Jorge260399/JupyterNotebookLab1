# **Scikit-image**

```python
from skimage import io
img = io.imread("data/2017-11-26-02-56-43.jpg", as_gray = True)
from skimage.transform import rescale, resize, downscale_local_mean
rescaled_img = rescale(img, 1.0/4.0, anti_aliasing = True)
resize_img = resize(img, (360, 200))
downscale_img = downscale_local_mean(img, (4,3))
plt.imshow(downscale_img)
```

```python
from skimage.filters import roberts, sobel, scharr, prewitt
edge_roberts = roberts(img)
edge_sobel = sobel(img)
edge_scharr = scharr(img)
edge_prewitt = prewitt(img)
fig, axes = plt.subplots(nrows=2, ncols=2, sharex=True, sharey=True,
                         figsize=(8, 8))
ax = axes.ravel()

ax[0].imshow(img, cmap=plt.cm.gray)
ax[0].set_title('Imagen Original')

ax[1].imshow(edge_roberts, cmap=plt.cm.gray)
ax[1].set_title('Deteccion de borde Roberts')

ax[2].imshow(edge_sobel, cmap=plt.cm.gray)
ax[2].set_title('Sobel')

ax[3].imshow(edge_scharr, cmap=plt.cm.gray)
ax[3].set_title('Scharr')

for a in ax:
    a.axis('off')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from skimage.filters.rank import entropy
from skimage.morphology import disk
img = io.imread("data/2017-11-26-02-56-43.jpg", as_gray = True)
entropy_img = entropy(img, disk(3))
from skimage.filters import threshold_otsu
thresh = threshold_otsu(entropy_img) 
binary= entropy_img <=thresh
plt.imshow(binary, cmap=plt.cm.gray)
print("El porcentaje de region blanca es: ", (np.sum(binary == 1)*100)/(np.sum(binary == 0) + np.sum(binary == 1)))
```

```python
from skimage import io, img_as_ubyte
import matplotlib.pyplot as plt
import numpy as np
from scipy import ndimage

img = img_as_ubyte(io.imread("data/2017-11-26-02-56-43.jpg", as_gray=True))
uniform_filtered_img = ndimage.uniform_filter(img, size=9)
#plt.imshow(uniform_filtered_img)
blurred_img = ndimage.gaussian_filter(img, sigma=3)
#plt.imshow(blurred_img) 
median_img = ndimage.median_filter(img, 3)
#plt.imshow(median_img)

fig, axes = plt.subplots(nrows=2, ncols=2, sharex=True, sharey=True,
                         figsize=(8, 8))
ax = axes.ravel()

ax[0].imshow(img)
ax[0].set_title('Original image')

ax[1].imshow(uniform_filtered_img)
ax[1].set_title('Uniform Filtered')

ax[2].imshow(blurred_img)
ax[2].set_title('Blurred')

ax[3].imshow(median_img)
ax[3].set_title('Median')

for a in ax:
    a.axis('off')

plt.tight_layout()
plt.show()
```
